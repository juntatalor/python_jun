```python
test_split = otr.capitalize().replace('.', ' ').replace('\n', '')
test_split = test_split
list_words = test_split.split(' ')
list_words
```

```python
len(list_words)
```

```python
list_words.count('кутузов')
```

```python
# используем регулярки в жизни, но пока рано
symb = [',', '!', '(', ')', '—', '?']
for i, word in enumerate(list_words):
    for s in symb:
        if s in word:
            list_words[i] = word.replace(s, '')
```

```python
list_words
```

```python
' '.join(list_words)
```

```python
list_words_without_empty = [word for word in list_words if word] 
```
